```python
class BankMachine:
    def __init__(self):
        self.secret_pin = ''
        self.__account_balance = 10000
    
    def get_balance(self):
        return self.__account_balance
    
    def set_balance(self,num):
        if type(num)==int:
            self.__account_balance=num
        else:
            print("please enter integer value only")

    def setup_pin(self):
        self.secret_pin = input('Set PIN: ')

    def view_balance(self):
        print('Balance:', self.__account_balance)

    def withdraw(self, amt):
        if self.secret_pin == input('PIN: '):
            if self.__account_balance >= amt:
                self.__account_balance -= amt  # CRASH if balance is a string!

machine = BankMachine()

# Villain corrupts the balance!
machine.account_balance = 'hehe'

machine.setup_pin()  # Works fine
machine.withdraw(500)
```

## Encapsulation

```python
class BankMachine:
    def __init__(self):
        self.__secret_pin = ''
        self.__account_balance = 10000
    
    # GETTER — read-only access to balance
    def retrieve_balance(self):
        return self.__account_balance
    
    # SETTER — write access WITH validation
    def assign_balance(self, new_amount):
        if type(new_amount) == int:
            self.__account_balance = new_amount
        else:
            print('Invalid! Balance must be an integer.')

    def setup_pin(self):
        self.__secret_pin = input('Set PIN: ')

    def view_balance(self):
        print('Balance:', self.__account_balance)

    def withdraw(self, amt):
        if self.__secret_pin == input('PIN: '):
            if self.__account_balance >= amt:
                self.__account_balance -= amt  # CRASH if balance is a string!
```

```python
machine = BankMachine()

# Villain corrupts the balance!
#machine.__account_balance = 'hehe'

machine.setup_pin()  # Works fine
machine.withdraw(500)
# CRASH! Can't do 'hehe' - 500
# TypeError: unsupported operand
# type(s) for -=: 'str' and 'int'
```

```python
class BankMachine:

    # --- Static variable ---
    cust_counter = 1
    a=100
    b=200

    def __init__(self):
        self.__cust_id = BankMachine.cust_counter
        BankMachine.cust_counter += 1
        self.__secret_pin = ''
        self.__account_balance = 0
        self.c= 100
        self.d= 200

    # --- Getter ---
    def retrieve_balance(self):
        return self.__account_balance

    # --- Setter with validation ---
    def assign_balance(self, new_amount):
        if type(new_amount) == int:
            self.__account_balance = new_amount
        else:
            print('Invalid: must be integer')
    
    def dispaly1(self):
        print("value of c is",self.c,self.d,self.__secret_pin,self.__account_balance,self.__cust_id)
        print("value of a is :",BankMachine.a)

    @staticmethod    
    def display():
        print("this is static method")
        print("value of a is :",BankMachine.a)

    # --- Static method ---
    @staticmethod
    def get_total_customers():
        return BankMachine.cust_counter - 1
```
