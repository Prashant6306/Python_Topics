```python
class Student:
    def __init__(self, full_name, gpa):
        self.full_name = full_name
        self.gpa = gpa
    def display(self):
        print("Full Name is:",self.full_name,"GPA is:",self.gpa)
```

```python
# --- Dict of Student objects ---
student_db = {
    'STU001': Student('Alice', 3.9),
    'STU002': Student('Bob',   3.5),
    'STU003': Student('Clara', 3.7),
}
```

```python
# --- Key lookup ---
print(student_db['STU001'])   # Alice
```

```python
for student_id, student_obj in student_db.items():
    print(f'ID: {student_id} | Name: {student_obj.full_name} | GPA: {student_obj.gpa}')
```

```python
class BankMachine:
    # Static variable — OUTSIDE __init__
    cust_counter = 1   # Shared by all instances

    def __init__(self,x="rajesh"):
        # Assign current count as this obj's ID
        self.cust_id  = BankMachine.cust_counter #instance variable
        self.name=x
        # Increment for next object
        BankMachine.cust_counter =BankMachine.cust_counter + 1

m1 = BankMachine()  # cust_id = 1
m2 = BankMachine("ganesh1")  # cust_id = 2
m3 = BankMachine("ganesh2")  # cust_id = 3

print(m1.cust_id)
print(m2.cust_id)
print(m3.cust_id)
# Works perfectly!
print(m1.name)
```

```python
class BankMachine:
    cust_counter = 1   # Static variable

    def __init__(self):
        self.cust_id = BankMachine.cust_counter
        BankMachine.cust_counter = BankMachine.cust_counter + 1
    
    def display(self):
        print(self.cust_id)

    @staticmethod
    def get_total_customers():   # no self!
        return BankMachine.cust_counter - 1

    @staticmethod
    def validate_pin_format(pin_str):   # utility
        return pin_str.isdigit() and len(pin_str) == 4

# --- Calling static methods ---
#m1 = BankMachine()
#m2 = BankMachine()

# Call via class name (recommended)
print(BankMachine.get_total_customers())
# Output: 2

print(BankMachine.validate_pin_format('4523'))  # True
print(BankMachine.validate_pin_format('ab12'))  # False
```
