```python
with open('diary.txt', 'r') as fobj:
    content = fobj.readlines()
    print(content[0:2])
# File closed automatically here
```

```python
# Safe writing — auto-close on exit
with open('diary1.txt', 'w') as fobj:
    fobj.write('Day 1: Started learning\n')
    fobj.write('Day 2: practiced more!\n')
    fobj.write("jai ganesh")
# File flushed and closed here
```

```python
# Read file line by line safely
with open('numbers.txt','r') as fobj:
    #print(fobj)
    #print(type(fobj))
    # result=fobj.read().strip()
    # print(result)
    for line in fobj:
        #print(line)
        val = line.strip()
        print(val)
# Handles large files efficiently
```

```python
import os

path = '/home/prashant/Desktop/AI_Engineer_with_MLops_Course/16. File input-output_8_Mar/file-input-output.ipynb'

# --- Existence checks ---
print(os.path.exists(path))     # True
print(os.path.isfile(path))     # True
print(os.path.isdir(path))      # False

# --- Path decomposition ---
print(os.path.dirname(path))
# -> /home/user/projects

print(os.path.basename(path))
# -> data.txt

print(os.path.split(path))
# -> ('/home/user/projects', 'data.txt')

print(os.path.splitext(path))
# -> ('/home/user/projects/data', '.txt')

# --- Building paths safely ---
base = '/home/user'
joined = os.path.join(base, 'docs', 'report.txt')
print(joined)   # /home/user/docs/report.txt

# --- Get file size ---
print(os.path.getsize(path), 'bytes')
```
