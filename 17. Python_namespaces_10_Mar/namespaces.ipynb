```python
def greet():
    msg = 'Hello!'   # 'msg' lives
    print(msg)       # in local NS

language = 'Python'  # lives in
version  = 3.11     # global NS
print(language)
greet()
```

```python
count = 100        # global

def show_count():
    count = 999    # local — shadows global!
    print(count)   # prints 999

show_count()
print(count)       # prints 100
```

```python
count = 100         # global

def update_count():
    global count    # tell Python: use global!
    count = 999     # modifies global!
    print(count)    # 999

update_count()
print(count)        # 999 ✅
```

```python
city = 'Paris'
population = 2_200_000
country = 'France'

def city_info():
    district = 'Marais'    # local
    global population
    population += 50_000   # modify global

    print(locals())        # only local vars
    print(globals())       # global vars

print(locals())  # same as globals() here
print(globals())
city_info()
```

### Inner Functions — Functions Inside Functions

```python
# Outer function
def order_summary():
    item = 'Coffee'
    price = 4.50
    print('--- Order Summary ---')

    # Inner function
    def show_receipt():
        tax = price * 0.10
        total = price + tax
        print(f'Item:  {item}')
        print(f'Price: $' + '{price:.2f}')
        print(f'Tax:   $' + '{tax:.2f}')
        print(f'Total: $' + '{total:.2f}')

    show_receipt()  # called inside outer

order_summary()
# show_receipt() <- ERROR: not accessible!
```

```python
def fun1():
    print("outer function")
    def fun2():
        print("Inner function")
    fun2()


fun1()
```

```python
tax_rate = 0.18   # global

def calc_price():
    base = 100        # enclosing
    discount = 10     # enclosing

    def apply_tax():
        # reads from enclosing + global
        after_disc = base - discount
        final = after_disc * (1 + tax_rate)
        print(f'After discount: {after_disc}')
        print(f'Final price:    {final:.2f}')

    apply_tax()  # call inner fn

calc_price()
```
