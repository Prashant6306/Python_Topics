### Pass By Reference

```python
class Employee:
    def __init__(self, emp_name, role):
        self.emp_name = emp_name
        self.role = role

# Function receives an Employee object
def display_info(emp):
    print('Inside fn, id:', id(emp))
    print(emp.emp_name, '-', emp.role)

# Function returns an Employee object
def build_employee():
    temp = Employee('Jamie', 'Manager')
    print(id(temp))
    return temp

# --- Proof: same id inside & outside ---
worker = Employee('Alex', 'Dev')
print('Outside fn, id:', id(worker))
display_info(worker)
# Output: SAME id in both lines!
```

```python
worker2=build_employee()
```

### Object Mutability

```python
class Employee:
    def __init__(self, emp_name, role):
        self.emp_name = emp_name
        self.role = role
        
def update_name(emp):
    emp.emp_name = 'Morgan'
    return emp

worker = Employee('Alex', 'Dev')
print(id(worker))   # e.g. 140234567890

result = update_name(worker)
print(id(result))   # SAME: 140234567890
# id unchanged → object was modified IN PLACE → MUTABLE!
```

## Encapsulation

```python
class BankMachine:
    def __init__(self):
        self.__secret_pin = ''
        self.__account_balance = 10000
    
    # GETTER — read-only access to balance
    def retrieve_balance(self):
        return self.__account_balance
    
    # SETTER — write access WITH validation
    def assign_balance(self, new_amount):
        if type(new_amount) == int:
            self.__account_balance = new_amount
        else:
            print('Invalid! Balance must be an integer.')

    def setup_pin(self):
        self.__secret_pin = input('Set PIN: ')

    def view_balance(self):
        print('Balance:', self.__account_balance)

    def withdraw(self, amt):
        if self.__secret_pin == input('PIN: '):
            if self.__account_balance >= amt:
                self.__account_balance -= amt  # CRASH if balance is a string!
```

```python
machine = BankMachine()

# Villain corrupts the balance!
#machine.__account_balance = 'hehe'

machine.setup_pin()  # Works fine
machine.withdraw(500)
# CRASH! Can't do 'hehe' - 500
# TypeError: unsupported operand
# type(s) for -=: 'str' and 'int'
```
